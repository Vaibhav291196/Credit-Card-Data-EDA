 In case age is less than 18, replace it with mean of age values.

b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit. (customer’s limit provided in acquisition table is the per transaction limit on his card),

```python
customer_spend= customer.merge(cr_spend,how='right',on='Customer')
customer_spend
```

```python
customer_spend.rename({'Amount':'spent_amount'},axis=1,inplace=True)
customer_spend
```

```python
customer_spend['spent_amount']=np.where(customer_spend['spent_amount']>customer_spend['Limit'],0.5*customer_spend['Limit'],customer_spend['spent_amount'])
customer_spend.head(25)
```

```python
def func(row):
  if row['spent_amount']>row['Limit']:
    row['spent_amount']=0.5*row['Limit']
  return row
```

```python
cr=customer_spend[['spent_amount','Limit']].apply(func,axis =1)
cr
```

```python
customer_repayment=cr_repayment.merge(customer,how='left',on='Customer')
customer_repayment
```

```python
customer_repayment.rename({'Amount':'repayment_amount'},axis=1,inplace=True)
customer_repayment
```

```python
def func(y):
  if y['repayment_amount']>y['Limit']:
    y['repayment_amount']=y['Limit']
  else:
    y['repayment_amount']>y['repayment_amount']
  return y
```

```python
customer_repayment=customer_repayment.apply(func,axis=1)
customer_repayment
```

What is the average monthly repayment by customers?

```python
customer_repayment.groupby('Month')['repayment_amount'].mean()
```

```python
cr_transactions=c1.merge(c2,how='inner',on=('Year','Month'))
cr_transactions
```

```python
cr_transactions['Monthly Profit']=cr_transactions['repayment_amount']-cr_transactions['spent_amount']

cr_transactions
```

```python
def func(x):
  if x['Monthly Profit']>0:
    x['Monthly_Earnings']=x['Monthly Profit']*0.029
  else:
    x['Monthly_Earnings']=0
  return x
```

```python
cr_transactions=cr_transactions.apply(func,axis=1)
cr_transactions.sort_values('Monthly_Earnings')
```

```python
earnings=cr_transactions[cr_transactions['Monthly_Earnings']>0]
earnings
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=earnings, x='Month',y='Monthly_Earnings', hue='Year',color='cornflowerblue')
plt.show()
```

Monthly earnings

```python
earnings_monthly=pd.DataFrame(earnings.groupby('Month')['Monthly_Earnings'].sum())
earnings_monthly
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=earnings_monthly,x='Month',y='Monthly_Earnings',color='brown')
plt.show()
```
